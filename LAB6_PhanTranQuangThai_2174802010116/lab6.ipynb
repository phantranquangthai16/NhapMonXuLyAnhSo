```python
#bài2
import cv2
import numpy as np

img = cv2.imread('exercise/colorful-ripe-tropical-fruits.jpg')

if img is None:
    print("Không tìm thấy ảnh.")
    exit()

result = img.copy()

x1, y1, w1, h1 = 470, 350, 200, 200
papaya = result[y1:y1+h1, x1:x1+w1]

x2, y2, w2, h2 = 1800, 200, 130, 140
watermelon = result[y2:y2+h2, x2:x2+w2]

def change_color(bgr_crop):
    hsv = cv2.cvtColor(bgr_crop, cv2.COLOR_BGR2HSV)
    hsv[..., 0] = (hsv[..., 0] + 90) % 180  
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

papaya_changed = change_color(papaya)
watermelon_changed = change_color(watermelon)

result[y1:y1+h1, x1:x1+w1] = papaya_changed
result[y2:y2+h2, x2:x2+w2] = watermelon_changed

cv2.imshow("Doi mau du du va dua hau", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#bài 3
import cv2
import numpy as np
import os

def rotate_image(image, angle):
    
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])
    new_w = int((h * sin) + (w * cos))
    new_h = int((h * cos) + (w * sin))
    M[0, 2] += (new_w / 2) - center[0]
    M[1, 2] += (new_h / 2) - center[1]
    rotated = cv2.warpAffine(image, M, (new_w, new_h), borderValue=(255, 255, 255))
    return rotated

def main():
    input_path = "exercise/quang_ninh.jpg"
    if not os.path.exists(input_path):
        print("Không tìm thấy ảnh!")
        return

    image = cv2.imread(input_path)

    mountain_roi = image[272:299, 100:400] 
    boat_roi = image[111:203, 200:350]     

    mountain_rotated = rotate_image(mountain_roi, 45)
    boat_rotated = rotate_image(boat_roi, 45)

    cv2.imwrite("mountain_rotated.jpg", mountain_rotated)
    cv2.imwrite("boat_rotated.jpg", boat_rotated)
    print("Đã lưu ảnh xoay 45 độ!")

if __name__ == "__main__":
    main()
```

```python
#bài 5
import cv2
import os
import numpy as np

def choose_image():
    image_folder = "exercise"
    images = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

    if not images:
        print("Không có ảnh trong thư mục.")
        return None

    print("Chọn ảnh muốn xử lý:")
    for idx, name in enumerate(images):
        print(f"{idx+1}. {name}")
    
    choice = int(input("Nhập số thứ tự ảnh: ")) - 1
    if 0 <= choice < len(images):
        return os.path.join(image_folder, images[choice])
    else:
        print("Lựa chọn không hợp lệ.")
        return None

def translate_image(img):
    dx = int(input("Nhập số pixel tịnh tiến theo trục x: "))
    dy = int(input("Nhập số pixel tịnh tiến theo trục y: "))
    M = np.float32([[1, 0, dx], [0, 1, dy]])
    translated = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
    return translated

def rotate_image(img):
    angle = float(input("Nhập góc xoay (độ): "))
    center = (img.shape[1] // 2, img.shape[0] // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
    return rotated

def scale_image(img, scale_factor):
    new_size = (int(img.shape[1] * scale_factor), int(img.shape[0] * scale_factor))
    scaled = cv2.resize(img, new_size, interpolation=cv2.INTER_LINEAR)
    return scaled

def coordinate_map(img):
    clone = img.copy()

    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            print(f"Tọa độ: ({x}, {y})")
            cv2.circle(clone, (x, y), 5, (0, 0, 255), -1)
            cv2.imshow("Coordinate Map", clone)

    cv2.imshow("Coordinate Map", clone)
    cv2.setMouseCallback("Coordinate Map", click_event)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def main():
    while True:
        print("\n--- MENU ---")
        print("T - Tịnh tiến")
        print("X - Xoay")
        print("P - Phóng to")
        print("H - Thu nhỏ")
        print("C - Coordinate Map")
        print("Q - Thoát")
        choice = input("Chọn thao tác (T/X/P/H/C/Q): ").strip().upper()

        if choice == "Q":
            break

        image_path = choose_image()
        if image_path is None:
            continue

        img = cv2.imread(image_path)
        if img is None:
            print("Không thể đọc ảnh.")
            continue

        result = None

        if choice == "T":
            result = translate_image(img)
        elif choice == "X":
            result = rotate_image(img)
        elif choice == "P":
            scale = float(input("Nhập hệ số phóng to (ví dụ 2.0): "))
            result = scale_image(img, scale)
        elif choice == "H":
            scale = float(input("Nhập hệ số thu nhỏ (ví dụ 0.5): "))
            result = scale_image(img, scale)
        elif choice == "C":
            coordinate_map(img)
            continue
        else:
            print("Lựa chọn không hợp lệ.")
            continue

        if result is not None:
            cv2.imshow("Kết quả", result)
            save_path = image_path.replace(".jpg", f"_{choice}.jpg").replace(".png", f"_{choice}.png")
            cv2.imwrite(save_path, result)
            print(f"Đã lưu ảnh kết quả: {save_path}")
            cv2.waitKey(0)
            cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
